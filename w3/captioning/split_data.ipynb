```python
import numpy as np
import random
from transformers import ResNetModel
from torch import nn
from torch.utils.data import Dataset
from PIL import Image
import torch.optim as optim
from torchvision.transforms import v2
import torch
import pandas as pd
import evaluate
import os
from torch.utils.data import DataLoader
import wandb
from sklearn.model_selection import train_test_split
```

```python
img_path = '/ghome/c5mcv04/MCV-C5-2025-Team4/dataset/foodDataset/Food_Images'
cap_path = '/ghome/c5mcv04/MCV-C5-2025-Team4/w3/data/f_cleaned.csv'

data = pd.read_csv(cap_path)
data["Image_Name"] += ".jpg"
valid_images = {img for img in os.listdir(img_path)}  
data = data[data["Image_Name"].isin(valid_images)].reset_index(drop=True)

image_filenames = data['Image_Name'].values
labels = data['Title'].values  

train_files, test_files, train_labels, test_labels = train_test_split(image_filenames, labels, test_size=0.2, random_state=42)
valid_files, test_files, valid_labels, test_labels = train_test_split(test_files, test_labels, test_size=0.5, random_state=42)

def load_image(img_filename):
    print(img_filename)
    img_path_full = os.path.join(img_path, img_filename)  
    img = Image.open(img_path_full).convert("RGB")  
    img = np.array(img)  
    return img_path_full

train_data = [(load_image(img), label) for img, label in zip(train_files, train_labels)]
valid_data = [(load_image(img), label) for img, label in zip(valid_files, valid_labels)]
test_data = [(load_image(img), label) for img, label in zip(test_files, test_labels)]

all_data = {
    'train': train_data,
    'valid': valid_data,
    'test': test_data
}

np.save('/ghome/c5mcv04/MCV-C5-2025-Team4/w3/captioning/food_data.npy', all_data)

print("Data has been saved to 'food_data_2.npy'.")
```
